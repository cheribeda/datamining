```python
import pandas as pd
```

```python
# Load the data from the CSV file

df=pd.read_csv("/Users/cheribeda/Desktop/Video_Games.csv")
```

```python
# Display first 10 columns

df.head(10)
```

```python
# Get the dimensions (number of rows and columns)
num_rows, num_cols = df.shape

# Display the dimensions
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
# Sort the DataFrame by critic score in descending order
top_games = df.sort_values(by='Critic_Score', ascending=False)

# Select the top five games
top_five_games = top_games.head(5)

# Display the top five games
print(top_five_games)
```

```python
# Count the number of video games in each genre
genre_counts = df['Genre'].value_counts()

# Display the number of video games in each genre
print(genre_counts)
```

```python
# Filter the DataFrame for games on the SNES platform
snes_games = df[df['Platform'] == 'SNES']

# Select the first five games
first_five_snes_games = snes_games.head(5)

# Display the first five games on the SNES platform
print(first_five_snes_games)
```

```python
# Group the data by publisher and sum the global sales for each publisher
publisher_sales = df.groupby('Publisher')['Global_Sales'].sum()

# Sort the publishers by total global sales in descending order
publisher_sales_sorted = publisher_sales.sort_values(ascending=False)

# Select the top five publishers
top_five_publishers = publisher_sales_sorted.head(5)

# Display the top five publishers with the highest total global sales
print(top_five_publishers)
```

```python
# Calculate the percentage of global sales from North America
df['NA_Sales_Percentage'] = (df['NA_Sales'] / df['Global_Sales']) * 100

# Display the first five rows of the new DataFrame
print(df.head(5))
```

```python
# Count the number of NaN entries in each column
nan_counts = df.isna().sum()

# Display the number of NaN entries in each column
print(nan_counts)
```

```python
# Convert non-numeric entries in the user score column to NaN
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')

# Calculate the median of user scores
median_user_score = df['User_Score'].median()

# Replace NaN entries in the user score column with the median value
df['User_Score'].fillna(median_user_score, inplace=True)

# Display the updated DataFrame
print(df)
```
